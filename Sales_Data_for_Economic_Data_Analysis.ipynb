######  Pertanyaan 1:

- Bagaimana tren penjualan secara keseluruhan?

```python
plt.figure(figsize = (18,4))

plt.subplot(1,2,1)
plt.hist(df["Profit"], color = "lightgreen")
plt.title("Profit Histogram")

plt.subplot(1,2,2)
sns.boxplot(data = df, x = df["Profit"], color = "lightgreen")
plt.title("Profit Boxplot")

plt.show()
```

```python
df.groupby('Year')['Profit'].sum().reset_index().sort_values("Profit")
```

```python
import matplotlib.pyplot as plt

# Mengelompokkan berdasarkan 'Year' dan menghitung jumlah 'Profit'
yearly_profit = df.groupby('Year')['Profit'].sum().reset_index()

# Mengurutkan hasil berdasarkan 'Profit' secara menurun
yearly_profit_sorted = yearly_profit.sort_values(by="Profit")

# Konversi kolom 'Year' menjadi string
yearly_profit_sorted['Year'] = yearly_profit_sorted['Year'].astype(str)

# Mengelompokkan berdasarkan 'Year' dan menghitung jumlah 'Sale'
yearly_sale = df.groupby('Year')['Sale'].sum().reset_index()

# Mengurutkan hasil berdasarkan 'Sale' secara menurun
yearly_sale_sorted = yearly_sale.sort_values(by="Sale")

# Konversi kolom 'Year' menjadi string
yearly_sale_sorted['Year'] = yearly_sale_sorted['Year'].astype(str)

# Membuat subplot dengan 1 baris dan 2 kolom
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Visualisasi pertama: Total Profit per Year
bars1 = axs[0].bar(yearly_profit_sorted['Year'], yearly_profit_sorted['Profit'], color='skyblue')

axs[0].set_title('Total Profit per Year')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Total Profit')
axs[0].grid(False)

# Menambahkan nilai di atas setiap batang
for bar in bars1:
    height = bar.get_height()
    axs[0].text(bar.get_x() + bar.get_width()/2, height, '{:,.0f}'.format(height), ha='center', va='bottom')

# Visualisasi kedua: Total Sale per Year
bars2 = axs[1].bar(yearly_sale_sorted['Year'], yearly_sale_sorted['Sale'], color='lightgreen')

axs[1].set_title('Total Sale per Year')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Sale')
axs[1].grid(False)

# Menambahkan nilai di atas setiap batang
for bar in bars2:
    height = bar.get_height()
    axs[1].text(bar.get_x() + bar.get_width()/2, height, '{:,.0f}'.format(height), ha='center', va='bottom')

# Menghapus label sumbu y
axs[0].set_yticklabels([])
axs[1].set_yticklabels([])

plt.tight_layout()
plt.show()
```

  - negara Manakah teratas berdasarkan penjualan dan profit?

```python
# Mengelompokkan berdasarkan 'Country' dan menghitung jumlah 'Profit'
country_profit = df.groupby('Country')['Profit'].sum().reset_index()

# Mengurutkan hasil berdasarkan 'Profit' secara menurun
country_profit_sorted = country_profit.sort_values(by="Profit", ascending=False)

# Visualisasi grafik batang untuk total profit per negara
plt.figure(figsize=(15, 6))

# Visualisasi kiri
plt.subplot(1, 2, 1)
bars = plt.bar(country_profit_sorted['Country'], country_profit_sorted['Profit'], color='skyblue')

# Menambahkan label nilai di atas setiap batang
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, "{:,.0f}".format(yval), va='bottom', ha='center')

plt.title('Total Profit per Country')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.ylabel('Total Profit')
plt.gca().set_yticks([])  # Menghapus label sumbu y

# Mengelompokkan berdasarkan 'Country' dan menghitung jumlah 'Sale'
country_sale = df.groupby('Country')['Sale'].sum().reset_index()

# Mengurutkan hasil berdasarkan 'Sale' secara menurun
country_sale_sorted = country_sale.sort_values(by="Sale", ascending=False)

# Visualisasi grafik batang untuk total sale per negara
plt.subplot(1, 2, 2)
bars = plt.bar(country_sale_sorted['Country'], country_sale_sorted['Sale'], color='lightgreen')

# Menambahkan label nilai di atas setiap batang
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, "{:,.0f}".format(yval), va='bottom', ha='center')

plt.title('Total Sale per Country')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.ylabel('Total Sale')
plt.gca().set_yticks([])  # Menghapus label sumbu y

plt.tight_layout()
plt.show()
```
